```python
# Standarization function 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_crm1_df = scaler.fit_transform(crm1.iloc[:,1:])
```

```python
scaled_crm1_df
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4,random_state=0)
kmeans.fit(scaled_crm1_df)
```

```python
crm1["Cluster"] = kmeans.predict(scaled_crm1_df)
```

```python
crm1
```

```python
kmeans.inertia_ #Within a cluster sum 0f square 
```

```python
wcss = []
for i in range(2,11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaled_crm1_df)
    dist = kmeans.inertia_
    wcss.append(dist)
```

```python
wcss
```

```python
plt.plot(range(2,11),wcss)
plt.title("Elbow Method")
plt.xlabel("Number of cluster")
plt.ylabel("Within a cluster sum of square")
plt.grid()
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaled_crm1_df)
    wcss.append(kmeans.inertia_)
    if i > 2:
      break
```

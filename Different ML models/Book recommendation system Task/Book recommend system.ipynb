```python
def get_important_features(data):
    important_features = []
    for i in range(data.shape[0]):
        a = (data["Book_Title"][i]+' '+data["Book.Rating"][i])
        important_features.append(a)
        
    return important_features
```

```python
#Convert text into matrix

from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cm = CountVectorizer().fit_transform(df["Book_Title"])
```

```python
cm
```

```python
cm.data.shape
```

```python
cs = cosine_similarity(cm)
```

```python
cs.shape
```

```python
cs
```

```python
new_book = "Goodbye to the Buttermilk Sky"

book = df[df.Book_Title==new_book].index.values[0]
```

```python
df[df["Book_Title"]=="Goodbye to the Buttermilk Sky"]
```

```python
df["Book_Title"]
```

```python
score = list(enumerate(cs[book]))
```

```python
score
```

```python
for i in score:
    print(i[1])
```

```python
sorted_scores = sorted(score,key=lambda x:x[1],reverse =True)
```

```python
sorted_scores
```

```python
sorted_scores[1:]
```

```python
j = 0
print("The 7 most recommended books to ",new_book,"are:\n")

for item in sorted_scores[1:]:
    Book_Title = df[df.index ==item[0]]['Book_Title'].values[0]
    print(j+1,Book_Title)
    j = j+1
    if j>6:
        break
```

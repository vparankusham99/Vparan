```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
crmD = pd.read_csv(r"C:\Users\Muralidhar\Downloads\crime_data.csv")
crmD
```

```python
crmD.drop(['Unnamed: 0'],axis=1,inplace=True)
crmD
```

```python
#Standarization (z-score)
from sklearn.preprocessing import StandardScaler
```

```python
stscaler = StandardScaler()
X = stscaler.fit_transform(crmD)
```

```python
pd.DataFrame(X)
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps=1, min_samples=7)
dbscan.fit(X)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
crmD["Label"] = dbscan.labels_
```

```python
crmD
```

```python
crmD["Label"].value_counts()
```
